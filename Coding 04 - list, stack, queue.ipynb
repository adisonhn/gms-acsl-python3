# List

```python
# Python list data structure
days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

# You can iterate over each element in a list using for loop
for day in days:
  print(day)
```

```python
print(type(days))

# you can use len() function to tell how long is a list
print(len(days))
```

```python
# you can use indexing [...] to access particular element in the list (starting from 0)
print(days[0])
print(days[1])
print(days[2])
print(days[3])
print(days[4])
print(days[5])
print(days[6])
```

```python
# or you can iterate using its index
for i in range(len(days)):
  print(days[i])
```

```python
# This will give an error saying:
#    index is out of range
# because the index 7 is out of range of accepted values 0 to 6
print(days[7])
```

```python
# you can use negative index
print(days[-1])  # gives the last element of the list
```

```python
print(days[-2])  # gives second to last element of the list
```

```python
# You can also get the last element of the list using len()
print(days[len(days)-1]) # days[6]
# And the second to last element of the list
print(days[len(days)-2]) # days[5]
```

```python
# Python list is heterogeneous, i.e. it can hold different data types
mixed_bag = [1.5, 2, 'Monday', False]
print(mixed_bag)
print(len(mixed_bag))
```

```python
# It can even hold lists
my_list = [[1], [1, 2], [1, 2, 3]]
print(len(my_list))

# you can use indexing [...] to access a particular 
print(my_list[0])
print(my_list[1])
print(my_list[2])
```

```python
# You can append items to a list
my_list = []
my_list.append('a')
my_list.append('b')
print(my_list)
print(len(my_list))
```

```python
names = ['Alice', 'Bob', 'Charlie', 'Bob', 'Dave']

names.remove('Alice')
print(names)
```

```python
names = ['Alice', 'Bob', 'Charlie', 'Bob', 'Dave']

names.remove('John')  # this will raise an error since there is no John
```

```python
names = ['Alice', 'Bob', 'Charlie', 'Bob', 'Dave']

# it's safer to check if John is in the names before attempting to remove John
if 'John' in names:
  names.remove('John')
```

```python
names = ['Alice', 'Bob', 'Charlie', 'Bob', 'Dave']

# Notice there are 2 Bobs in this list.
# removing Bob will just remove the first occurrence of it
names.remove('Bob')
print(names)
# will print: ['Alice', 'Charlie', 'Bob', 'Dave']
```

```python
names = ['Alice', 'Bob', 'Charlie', 'Bob', 'Dave']

# How to remove all Bobs?
# We can keep removing Bob while it can still be found in the list
while 'Bob' in names:
  names.remove('Bob')

print(names)
# will print: ['Alice', 'Charlie', 'Dave']
```

```python
names = ['Alice', 'Bob', 'Charlie', 'Bob', 'Dave']

# You can also remove by index, rather than by its value
del names[2] # this will delete Charlie
print(names)
# will print: ['Alice', 'Bob', 'Bob', 'Dave']
```

# Using List as Stack

Stack data structure is also known as LIFO (Last In First Out). The element that you put in the last will be the first one to be removed when you need to take one out.

When you append an item, the item will be added to the end of the list. This is equivalent to pushing an element to a stack.

When you pop an item, the last item will be removed from the list.



```python
my_stack = []

my_stack.append(10)
my_stack.append(20)
my_stack.append(30)
my_stack.append(40)
print(my_stack)
# will print: [10, 20, 30, 40]

x = my_stack.pop()
print(x)
# will print: 40
print(my_stack)
# will print: [10, 20, 30]

x = my_stack.pop()
print(x)
# will print: 30
print(my_stack)
# will print: [10, 20]
```

# Using List as Queue

Queue data structure is also known as FIFO (First In First Out). The element that goes in first will be the one that will go out first.

When you append an item, the item will be added to the end of the list. This is equivalent to pushing an element to a stack.

When you pop an item, you will call `pop(0)` so that the first item in the list will be removed.



```python
my_queue = []

my_queue.append(10)
my_queue.append(20)
my_queue.append(30)
my_queue.append(40) 
print(my_queue)
# will print: [10, 20, 30, 40]

x = my_queue.pop(0) # instead of just pop()
print(x)
# will print: 10
print(my_queue)
# will print: [20, 30, 40]

x = my_queue.pop(0)
print(x)
# will print: 20
print(my_queue)
# will print: [30, 40]
```

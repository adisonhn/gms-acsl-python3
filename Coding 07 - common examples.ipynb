```python
N = 4
# Repeat N times
for i in range(N):
  # write what you want to repeat here, for ex:
  print(i)
# This will print:
# 0
# 1
# 2
# 3
```

```python
name = 'ACSL'
# Repeat for every character in a string
for i in name:
  print(i)
# This will print:
# A
# C
# S
# L
```

```python
cities = ['Austin', 'Dallas', 'Houston', 'San Antonio']
# Repeat for every item in a list
for city in cities:
  print(city)
# This will print:
# Austin
# Dallas
# Houston
# San Antonio
```

```python
# Read an input line that contains 2 integers separated by space 
# into variables a, b as integers
# Example input:
# 12 345
a, b = input().split()
a = int(a)
b = int(b)

# Read an input line that contains 2 integers separated by comma
# into variables a, b as integers
# Example input:
# 12,345
a, b = input().split(',') # a, b are strings
a, b = int(a), int(b)     # a, b are now converted to ints

# Note:
a = int(a)
b = int(b)
# can also be written as:
a, b = int(a), int(b)
```

```python
# Normally print() will print and go to next line
print('A')
print('B')
print('C')
# will print in 3 lines: 
# A
# B

# You can specify additional parameter end='' to not go to next line
print('A', end='')
print('B', end='')
print('C', end='')
# will print in 1 line:
# ABC

# What happens if you let end='...'?
print('A', end='...')
print('B', end='...')
print('C', end='...')
# will print in 1 line:
# A...B...C...
```

```python
# We can use X in a_string to see if a_string contains X
# Example:
name = 'Grisham'
print('i' in name) # will print: True
print('ish' in name) # will print: True
print('sam' in name) # will print: False

# We can also use X in a_list to see if a_list contains X
# Example:
mylist = [4, 'abc', -1, False]
print(-1 in mylist) # will print: True
print('abc' in mylist) # will print: True
print(True in mylist) # will print: False
print('4' in mylist) # will print: False
```

```python
# Example: count number of vowels in a string
mystring = 'Grisham'
vowel_count = 0
for c in mystring:
  if c in 'aeiou':
    vowel_count += 1
print(vowel_count) # will print: 2

# This expression:
#   c in 'aeiou'
# can also be written as:
#   c in ['a', 'e', 'i', 'o', 'u']
```

```python
n = 5

# String concatenation using operator +
# Notice how each operand must be of type string
print(str(n) + ' squared is ' + str(n**2))
# will print: 5 squared is 25

# Python3 supports what's called f-string (or formatted string)
# using f'...' (or f"...")
# It's called f-string because you put f before the string
# Notice that we do not need to convert numbers to string
print(f'{n} squared is {n**2}')
# will print: 5 squared is 25
```

```python
# Multiplying a string with a number
print('$' * 3) # will print: $$$
print('ab' * 3) # will print: ababab

# Multiplying a list with a number
print([0] * 3) # will print: [0, 0, 0]
print([1, 2] * 3) # will print: [1, 2, 1, 2, 1, 2]
```

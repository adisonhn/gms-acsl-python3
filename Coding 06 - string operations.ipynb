```python
name = 'Grisham'

# You can use for loop to iterate over every character in a string
for i in name:
  print(i)
```

```python
# Length of the string
print(len(name))
```

```python
# Get the first character
print(name[0])

# And last character
print(name[-1])
# or you can use:
print(name[len(name)-1])
```

```python
# convert to upper case
print(name.upper())

# convert to lower case
print(name.lower())

# note: this doesn't change the value of name itself
print(name)
```

```python
# To change the variable name to contain uppercase letters, you need to reassign the uppercase value to the variable
name = name.upper()
print(name)
```

```python
# You can also get a substring of a string using indexing

# starts with index 1, and include everything all the way to the end
print(name[1:])

# starts with index 1, and include until index 3 (i.e. before index 4)
print(name[1:4])

# starts with index 1, and exclude the last character
print(name[1:-1])
```

```python
# String can also be split into a list of strings

sentence = 'String can also be split into a list of strings'
words = sentence.split(' ')
print(words)
# will print: ['String', 'can', 'also', 'be', 'split', 'into', 'a', 'list', 'of', 'strings']
```
